```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import scipy.io as io
```

## Load image and correspondences

```python
imL = cv2.imread("frame0.png")
imR = cv2.imread("frame1.png")
size = np.shape(imL)

correspondences = io.loadmat("Correspondences.mat")
```

# Estimate Fundamental Matrix

```python
def normalize_points(x):
    
    if (np.shape(x)[0]==2):
        x = np.concatenate((x, np.ones(shape=(1,np.shape(x)[1]))), axis=0)
        
        m = np.mean(x,axis=1)
                
        T = np.linalg.inv(np.array([[m[0] + m[1], 0, m[0]/2], [0, m[0] + m[1], m[1]/2], [0, 0, 1]]))
        
        x_normalized = T @ x
        x_normalized = x_normalized[0:2,:]
    
    return x_normalized, T

def assemble_y_matrix(pl, pr):   
                
    xl = pl[0]
    xr = pr[0]
    yl = pl[1]
    yr = pr[1]
                 
    y = np.column_stack((xr*xl, xr*yl, xr, yr*xl, yr*yl, yr, xl, yl, np.ones(shape=np.shape(pl)[1])))  
    return y

def estimate_fundamental_matrix(pixels_imageL, pixels_imageR):
    
    # normalize image coordinates
    [pixels_imageL_norm, TL] = normalize_points(pixels_imageL)
    [pixels_imageR_norm, TR] = normalize_points(pixels_imageR)
            
    # assemble Y matrix
    Y = assemble_y_matrix(pixels_imageL_norm, pixels_imageR_norm)
    
    # Solve linear system and enforce 2-rank constraint    
    U,S,V = np.linalg.svd(Y) 
    V = np.transpose(V)
    F_norm = np.reshape(V[:,-1],newshape=(3,3))
    uf,sf,vf = np.linalg.svd(F_norm)  
    vf = np.transpose(vf)
    
    F_norm_prime = uf @ np.diag(np.array([sf[0], sf[1], 0])) @ np.transpose(vf)    
    
    # de-normalize F
    F = np.transpose(TR) @ F_norm_prime @ TL;
    return F
    
F = estimate_fundamental_matrix(correspondences['pixelsImageL'], correspondences['pixelsImageR'])
```

```python
# Show an example of epipolar line computed with the fundamental matrix
aPixelOnLeftImage = [164, 117, 1]
lineCoeff = F @ aPixelOnLeftImage
x = np.arange(0,np.shape(imR)[1])
y = (-lineCoeff[0]/lineCoeff[1])*x - (lineCoeff[2]/lineCoeff[1]);

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(imL)
plt.plot(aPixelOnLeftImage[0], aPixelOnLeftImage[1], '.r')

plt.subplot(1,2,2)
plt.imshow(imR)
plt.plot(x,y,'b')
```

```python
def ncc(template, sample):
    
    if(np.shape(template) != np.shape(sample)):
        return -1
    
    mu_t = np.mean(template)
    mu_s = np.mean(sample)
    var_t = np.var(template)
    var_s = np.mean(sample)
    
    ncc = 0
    for y in range(0, np.shape(template)[0]):
        for x in range(0, np.shape(template)[1]):
            
            ncc += (template[y,x] - mu_t)*(sample[y,x] - mu_s)/(var_t*var_s)
    
    return ncc/(np.shape(template)[0]*np.shape(template)[1])
    
```
