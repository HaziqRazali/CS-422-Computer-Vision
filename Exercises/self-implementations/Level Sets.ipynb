```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read image and convert to grayscale

```python
im = cv2.imread("shape.png")
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

Smooth the image to reduce noise
* also helps increase the stop region

```python
im = cv2.GaussianBlur(im, ksize=(21,21), sigmaX = 7)
plt.figure()
plt.imshow(im, cmap="gray")
```

Define the stopping function to be the norm of the gradient image

```python
def stopping_fun(im):
    
    # Compute gradient in x and y direction
    grad_x = cv2.Sobel(im, cv2.CV_64F, 1, 0, 5)
    grad_y = cv2.Sobel(im, cv2.CV_64F, 0, 1, 5)
    
    # Compute gradient magnitude
    grad_mag = grad_x * grad_x + grad_y * grad_y
    
    return 1. / (1. + grad_mag)

F = stopping_fun(im)
plt.figure()
plt.imshow(F, cmap="gray")
```

Initialize surface phi

```python
def default_phi(x):
    # Initialize surface phi at the border (5px from the border) of the image
    # i.e. 1 outside the curve, and -1 inside the curve
    phi = np.ones(x.shape[:2])
    phi[5:-5, 5:-5] = -1.
    return phi

dt = 1
phi = default_phi(im)

for i in range(75):
    
    # Compute gradient in x and y direction
    dphi_x = cv2.Sobel(phi, cv2.CV_64F, 1, 0, 3)
    dphi_y = cv2.Sobel(phi, cv2.CV_64F, 0, 1, 3)    
    # Compute gradient magnitude
    dphi_norm = np.sqrt(dphi_x * dphi_x + dphi_y * dphi_y)

    phi = phi + F * dphi_norm

plt.imshow(im, cmap="gray")
plt.contour(phi, 0)
plt.show()
```
